```python
import yfinance as yf
import pandas as pd

# Download Brent Crude Oil price data
oil = yf.download("BZ=F", start="2015-01-01", end="2024-12-31")

# Keep only the closing price
oil = oil[["Close"]]

# Rename column
oil.columns = ["Price"]

# Check it worked
print(oil.shape)
print(oil.head())
print(oil.tail())
```

```python
import yfinance as yf
import pandas as pd
from datetime import date

# Download Brent Crude Oil price data up to today
oil = yf.download("BZ=F", start="2015-01-01", end=date.today().strftime("%Y-%m-%d"))

# Keep only the closing price
oil = oil[["Close"]]

# Rename column
oil.columns = ["Price"]

# Check it worked
print(f"Data runs from {oil.index.min()} to {oil.index.max()}")
print(f"Total trading days: {oil.shape[0]}")
print(oil.tail())
```

```python
# Save to our data folder
oil.to_csv("../data/oil_prices_raw.csv")

print("Data saved successfully!")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 5))
plt.plot(oil.index, oil["Price"], color="darkorange", linewidth=1)
plt.title("Brent Crude Oil Price 2015 - 2026", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
